## Nova abordagem

```python
import pandas as pd

# 1. Carregar o ficheiro original
# Certifique-se de que o ficheiro está na mesma pasta do script ou use o caminho completo
df = pd.read_csv('/workspaces/Neo4J_Recomendacao_Musica/scripts/archive/spotify_songs.csv')

# 2. Agrupar pelas 3 colunas e contar as ocorrências (totalizar)
# O drop_duplicates é feito implicitamente pelo groupby quando resumimos os dados
df.drop_duplicates(subset=['playlist_subgenre', 'playlist_genre', 'track_artist'])
df_sumarizado = df.groupby(['playlist_genre', 'playlist_subgenre']).size().reset_index(name='total_musicas')

# 3. Ordenar o resultado para facilitar a leitura (opcional)
df_sumarizado = df_sumarizado.sort_values(by=['playlist_genre', 'total_musicas'], ascending=[True, False])
```

```python
# 4. Guardar no novo ficheiro CSV
df_sumarizado.to_csv('/workspaces/Neo4J_Recomendacao_Musica/scripts/archive/spotify_resumo_totalizado.csv', index=False)

# Visualizar o resultado no terminal
print("Resumo gerado com sucesso:")
print(df_sumarizado.head(100))
```
